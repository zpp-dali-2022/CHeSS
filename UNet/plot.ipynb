```python

import pandas as pd
import matplotlib.pyplot as plt

# Load CSV file into a DataFrame
data = pd.read_csv('benchmark_results.csv')
data['benchmark'] = data['benchmark'].str.replace('_benchmark$', '', regex=True)

# Calculate the mean of the 'throughput' column
means = data.groupby('benchmark')['throughput'].mean()

means.plot(kind='bar', figsize=(10, 5), title='Throughput of UNet benchmarks')

```
