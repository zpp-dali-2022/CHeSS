```python

import pandas as pd
import matplotlib.pyplot as plt

# Load CSV file into a DataFrame
data = pd.read_csv('benchmark_results.csv')
data['benchmark'] = data['benchmark'].str.replace('_benchmark$', '', regex=True)
means = data.groupby('benchmark')['throughput'].mean()

ax = means.plot.bar(
    figsize=(8, 6), width =0.5, rot=0, title="Mean UNet training throughput",
    zorder=2, fontsize=20,
)
ax.set_title(ax.get_title(), fontsize=20, pad=20)
ax.set_xlabel('Rice compression algorithm', fontsize=20, labelpad=20)
ax.set_ylabel('throughput (images/s)', fontsize=20, labelpad=20)
ax.grid(axis='y', linestyle='--', color='gray', zorder=1, alpha=0.3)


plt.savefig('training_throughput.png', dpi=300, bbox_inches='tight')
```

```python
mean_time = data.groupby("benchmark")["time"].mean()
pd.DataFrame(mean_time)


ax = mean_time.plot.bar(
    figsize=(8, 6), width =0.5, rot=0, title="Mean UNet training time",
    zorder=2, fontsize=20,
)
# Hide the x-axis label

ax.set_title(ax.get_title(), fontsize=20, pad=20)
ax.set_xlabel('Rice compression algorithm', fontsize=20, labelpad=20)
ax.set_ylabel('time (s)', fontsize=20, labelpad=20)
ax.grid(axis='y', linestyle='--', color='gray', zorder=1, alpha=0.3)

plt.savefig('training_time.png', dpi=300, bbox_inches='tight')
```
